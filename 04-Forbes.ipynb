```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sys
sys.path.append("./forbes/")
from forbes import parse_top
```

```python
parse_top()
```

```python
df = pd.read_csv("./forbes/lines.csv", delimiter='|', names=['Rank','Name','Net Worth','Change','Age','Country/Territory','Source','Industry','control'])
```

```python
df.iloc[915:950, :]
```

```python
df['Net Worth Num'] = pd.to_numeric(df['Net Worth'].str.split(' ')[:].str[0].str[1:])
```

```python
df.tail()
```

```python
df.drop(columns='control').to_csv("./forbes/forbes.csv", index=False)
```

```python
df.drop(columns='control')
```

---
## Plots

```python
fig, ax = plt.subplots(figsize=(6,8))
sns.countplot(data=df, y='Country/Territory', order=list(df['Country/Territory'].value_counts().index), orient='v', ax=ax)

plt.show()
```

```python
df[['Country/Territory', 'Net Worth Num']].groupby('Country/Territory').sum().plot(kind='bar')
```

---
Adding continents

```python
cc = pd.read_csv("forbes/continent-country.txt")
fb = pd.read_csv("forbes/forbes.csv")
```

```python
fb2 = fb.merge(cc, how='left', left_on='Country/Territory', right_on='Country')
```

```python
fig, ax = plt.subplots(figsize=(6,8))
sns.countplot(data=fb2, y='Continent', order=list(fb2['Continent'].value_counts().index), orient='v', ax=ax)

plt.show()
```

```python
fb2.drop(columns=['Country']).to_csv("./forbes/forbes_continent.csv", index=False)
```
