# Forbes Billionaires

[Forbes Billionaire list](https://www.forbes.com/billionaires/)

```python
import pandas as pd
```

```python
f = open("./forbes/top.txt", "r")
i=0
a = []

for x in f:
    if ((i%8 != 0) & (i != 0)):
        a.append(x.replace('\n', '|'))
    elif (i%8 == 0):
        a.append('\n')
        a.append(x.replace('\n', '|'))
    i = i + 1
    
a[0] = ''
```

```python
q = ''.join(a)

w = open("./forbes/lines.csv", "w")
w.write(q)
w.close()
```

```python
df = pd.read_csv("./forbes/lines.csv", delimiter='|', names=['Rank','Name','Net Worth','Change','Age','Country/Territory','Source','Industry','control'])
```

```python
df['Net Worth Num'] = pd.to_numeric(df['Net Worth'].str.split(' ')[:].str[0].str[1:])
```

```python
df.drop(columns='control').to_csv("./forbes/forbes.csv", index=False)
```
